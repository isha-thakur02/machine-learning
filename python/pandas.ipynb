```python
import pandas as pd
student_records = pd.read_csv("student_records.csv")
student_marks = pd.read_csv("student_marks.csv")
student_records
```

```python
# 1. Merge both datasets on 'student_id'
merge = pd.merge(student_records, student_marks, on = "student_id") 
```

```python
merge
```

```python
# 2. Get students older than 20
student_records[student_records["age"]>20]
```

```python
# 3. Get top 5 students in math
student_marks.nlargest(5,"math_marks")
```

```python
# 4. Calculate average marks in Math and Science
student_marks[["math_marks","science_marks"]].mean()
```

```python
# 5. Find students scoring above 80 in both subjects
merge[(merge["math_marks"] > 80) & (merge["science_marks"] > 80)]
```
