```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

# Line Chart

```python
#  Plotting
x = np.arange(1,10)
y = np.arange(11,20)
plt.plot(x,y)
```

```python
# Add Grid
plt.grid()
plt.plot(x,y)
plt.grid(c = "blue", ls = "dotted", lw = 2)
```

```python
#  Assign Title
plt.title("Line Plot", c = "grey")
plt.plot(x,y)
```

```python
# Add Labels
plt.xlabel("Number upto 1-9")
plt.ylabel("Number upto 11-19")
plt.plot(x,y)
```

```python
#  Markers
# ms => marker size
# mec = > marker egde color
# mfc => marker face color

plt.plot(x,y, marker = "", ms = 10, mec = "black", mfc = "grey")
```

```python
# Line
plt.plot(x,y , ls ="dotted")
```

```python
values = np.arange(10,60,10)
plt.plot(values)
```

```python
#  Plot two or multiple lines
x = np.arange(1,10)
y = np.arange(11,20)
values = np.arange(21,30)
plt.plot(x,y)
plt.plot(x)
plt.plot(values,x)
```

```python
x = np.arange(25,60,5)
y = np.arange(35000,70000,5000)
plt.plot(x,y, marker = ".")
plt.grid()
plt.title("Age vs Income")
plt.xlabel("Age")
plt.ylabel("Income")
```
