```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# Legends
x = np.array([1,3,4,7,9])
y = np.array([5,7,6,10,12])

p = np.array([1,2,5,7,10])
q = np.array([6,8,7,9,11])

plt.plot(x,y, label = "first")
plt.plot(p,q, label = "second")
plt.legend()
```

# Scatter Plot

```python
x = np.array([12,33,24,35,17])
y = np.array([72,34,45,3,89])
plt.scatter(x,y)
```

# Bar chart

```python
categories = ["a", "b", "c", "d"]
values = [10,20,40,50]

plt.bar(categories, values, color = "silver")
```

```python
categories = ["a", "b", "c", "d"]
values = [10,20,40,50]

plt.barh(categories, values, color = "grey")
```

```python
categories = ["a", "b", "c", "d"]
values2023 = [10,20,40,50]
values2024 = [40,60,80,70]

plt.bar(categories,values2023, label = "2023")
plt.bar(categories,values2024, label = "2024", alpha = 0.1)
plt.legend()
```
