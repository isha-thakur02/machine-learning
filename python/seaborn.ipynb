```python
# Heatmap
import seaborn as sns
import numpy as np
a = np.linspace(1,10,20).reshape(4,5)
sns.heatmap(a)
```

```python
import pandas as pd
data = pd.read_csv("anagrams.csv")
x = data.drop(columns = "attnr")
# x
sns.heatmap(x, vmin = 0, vmax = 12, cmap = "cividis_r", cbar = False)
```

```python
#  Boxplot
data1 = sns.load_dataset("tips")
data1
sns.boxplot(x = "day", y = "total_bill", data = data1, hue = "time", color = "black")
```

```python
# Pairplot
data2 = sns.load_dataset("tips")
# data2
sns.pairplot(data2, hue = "smoker")
```
